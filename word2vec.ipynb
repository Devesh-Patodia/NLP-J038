```
import pandas as pd
import nltk
nltk.download('stopwords')
```

```
df= pd.read_json("http://snap.stanford.edu/data/amazon/productGraph/categoryFiles/reviews_Office_Products_5.json.gz" , lines=True)
df.head()
df = df.iloc[0:100,:]
```

```
from nltk.corpus import stopwords
stop = stopwords.words('english')
```

```
punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
def punctuation(str):
  no_punct = ""
  for char in str:
    if char not in punctuations:
      no_punct = no_punct + char
  return no_punct
```

```
df['reviewText'] = df['reviewText'].apply(punctuation)
```

```
df.head()
```

```
df['reviewText'] = df['reviewText'].str.lower().str.split()
```

```
df['reviewtext_no_stopwords'] = df['reviewText'].apply(lambda x: [item for item in x if item not in stop])
```

```
df.head()
```

```
def getUniqueWords(text) :
    uniqueWords = [] 
    for i in text:
        if not i in uniqueWords:
            uniqueWords.append(i)
    return uniqueWords
```

```
df['reviewText_unique_words'] = df['reviewtext_no_stopwords'].apply(getUniqueWords)
df.head()
```

```
from itertools import combinations
```

```
def combination(text):
  return(list(combinations(text,2)))
```

```
df['reviewText_combination'] = df['reviewText_unique_words'].apply(combination)
df.head()
```

```
print(df['reviewText_combination'][0])
```

```
from gensim.models import Word2Vec
import gensim
import warnings
warnings.filterwarnings("ignore")
```

```
model1 = gensim.models.Word2Vec(df['reviewText_unique_words'])
```
