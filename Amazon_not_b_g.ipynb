```
import pandas as pd
import gzip
import re
```

```
df = pd.read_json("http://snap.stanford.edu/data/amazon/productGraph/categoryFiles/reviews_Office_Products_5.json.gz" , lines=True)
df.head()
```

```
df.shape
```

```
def word_count(str):
    counts = dict()
    words = str.split()

    for word in words:
        if word in counts:
            counts[word] += 1
        else:
            counts[word] = 1

    return counts
```

```
df['reviewText_wordcount'] = df['reviewText'].apply(word_count)
df['summary_wordcount'] = df['summary'].apply(word_count)
df.head()
```

```
df.shape
```

```
vowels = ['a','e','i','o','u','A','E','I','O','U']
```

```
def split_string(str):
  str = re.sub('[^a-zA-Z]',' ', str)
  return list(str)
```

```
def letter(text):
  dict = {"vowels":0,
          "consonants":0}
  for word in text:
    words = split_string(word)
    for x in words:
      if x in vowels:
        dict["vowels"] += 1
      else:
         dict["consonants"] += 1
  return dict
```

```
df['reviewText_vowelcount'] = df['reviewText'].apply(lambda a:letter(a))
df['summary_vowelcount'] = df['summary'].apply(lambda b:letter(b))
df.head()
```

```
def words_begin_b(string):
    string = re.sub('[^a-zA-Z]',' ', string) 
    list_1 = re.findall(r'\b[bB]\w+', string)

    return(list_1)
```

```
df["reviewText_words_begin_b"] = df.reviewText.apply(words_begin_b)
df["summary_words_begin_b"] = df.summary.apply(words_begin_b)
df.head()
```

```
def words_end_g(string):
    string = re.sub('[^a-zA-Z]',' ', string) 
    list_2= re.findall(r'\w+[gG]\b', string) 

    return(list_2)
```

```
df["reviewText_words_end_g"] = df.reviewText.apply(words_end_g)
df["summary_words_end_g"] = df.summary.apply(words_end_g)
df.head()
```

```
def words_begin_end_b_g_both(string):
    string = re.sub('[^a-zA-Z]',' ', string) 
    list_3= re.findall(r'\b[bB]\w+[gG]\b', string) 

    return(list_3)
    
df["reviewText_BG"] = df.reviewText.apply(words_begin_end_b_g_both)
df["summary_BG"] = df.summary.apply(words_begin_end_b_g_both)
df.head()
```

```
def union_b_g(string):
    string = re.sub('[^a-zA-Z]',' ', string) 
    list4 = re.findall(r'\b[bB]\w+', string) 
    list5 = re.findall(r'\w+[gG]\b', string) 
    return (list(set().union(list4, list5)))

df["reviewText_union"] = df.reviewText.apply(union_b_g)
df["summary_union"] = df.summary.apply(union_b_g)
df.head()
```

```
punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
def punctuation(str):
  no_punct = ""
  for char in str:
    if char not in punctuations:
      no_punct = no_punct + char
  return no_punct
```

```
def no_end_g(string):
    result_1 = re.findall(r'\w+(?<!g)\b', string)       

    return (result_1)

df["reviewText_no_end_g"] = df['reviewText'].apply(punctuation)
df["reviewText_no_end_g"] = df['reviewText_no_end_g'].apply(no_end_g)
df["summary_no_end_g"] = df['summary'].apply(punctuation)
df["summary_no_end_g"] = df['summary_no_end_g'].apply(no_end_g)
df.head()
```

```
def no_begin_b(string):
    string = string.lower()
    result_2 = re.findall(r'\b(?!b)\w+', string)      
    return (result_2)

df["reviewText_no_begin_b"] = df['reviewText'].apply(punctuation)
df["reviewText_no_begin_b"] = df['reviewText_no_begin_b'].apply(no_begin_b)
df["summary_no_begin_b"] = df['summary'].apply(punctuation)
df["summary_no_begin_b"] = df['summary_no_begin_b'].apply(no_begin_b)
df.head()
```

```
def union_no_b_g(string):
    string = string.lower() 
    result_3 = re.findall(r'\b(?!b)\w+(?<!g)\b', string) 
    
    return (result_3)

df["reviewText_union_no_b_g"] = df['reviewText'].apply(punctuation)
df["reviewText_union_no_b_g"] = df['reviewText_union_no_b_g'].apply(union_no_b_g)
df["summary_union_no_b_g"] = df['summary'].apply(punctuation)
df["summary_union_no_b_g"] = df['summary_union_no_b_g'].apply(union_no_b_g)
df.head()
```
