```
import numpy as np
import pandas as pd
```

```
!wget https://rajpurkar.github.io/SQuAD-explorer/dataset/train-v2.0.json
```

```
df = pd.read_json('/content/train-v2.0.json')
```

```
df.head()
```

```
df.drop(df.tail(2).index,inplace=True)
```

```
df.shape
```

```
titles = []
paragraphs = []

for i in range(len(df)):
  titles.append(df['data'][i]['title'])
  paragraphs.append(df['data'][i]['paragraphs'])
```

```
df['titles'] = titles
df['paragraphs'] = paragraphs
df.head()
```

```
context = []
qas = []

for j in range(len(df)):
  context.append(df['paragraphs'][j][0]['context'])
  qas.append(df['paragraphs'][j][0]['qas'])
```

```
df['context'] = context
df['qas'] = qas
df.head()
```

```
answers = []
ids = []
is_impossible = []
questions = []

for k in range(len(df)):
  answers.append(df['qas'][k][0]['answers'])
  ids.append(df['qas'][k][0]['id'])
  is_impossible.append(df['qas'][k][0]['is_impossible'])
  questions.append(df['qas'][k][0]['question'])
```

```
df['answers'] = answers
df['ids'] = ids
df['is_impossible'] = is_impossible
df['questions'] = questions
df.head()
```

```
text = []
answer_start = []

for m in range(len(df)):
  text.append(df['answers'][m][0]['text'])
  answer_start.append(df['answers'][m][0]['answer_start'])
```

```
df['text'] = text
df['answer_start'] = answer_start
df.head()
```

```
data = df[['titles','context','answer_start','text','ids','is_impossible','questions']]
data.head()
```

```
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
```

```
model = tfidf.fit_transform(data['questions'])
```

```
model = pd.DataFrame(model.toarray(), columns=tfidf.get_feature_names())
```
