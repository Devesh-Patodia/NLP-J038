```
titles = []
paragraphs = []

for i in range(len(df)):
  titles.append(df['data'][i]['title'])
  paragraphs.append(df['data'][i]['paragraphs'])
```

```
df['titles'] = titles
df['paragraphs'] = paragraphs
df.head()
```

```
context = []
qas = []

for j in range(len(df)):
  context.append(df['paragraphs'][j][0]['context'])
  qas.append(df['paragraphs'][j][0]['qas'])
```

```
df['context'] = context
df['qas'] = qas
df.head()
```

```
answers = []
ids = []
is_impossible = []
questions = []

for k in range(len(df)):
  answers.append(df['qas'][k][0]['answers'])
  ids.append(df['qas'][k][0]['id'])
  is_impossible.append(df['qas'][k][0]['is_impossible'])
  questions.append(df['qas'][k][0]['question'])
```

```
df['answers'] = answers
df['ids'] = ids
df['is_impossible'] = is_impossible
df['questions'] = questions
df.head()
```

```
text = []
answer_start = []

for m in range(len(df)):
  text.append(df['answers'][m][0]['text'])
  answer_start.append(df['answers'][m][0]['answer_start'])
```

```
df['text'] = text
df['answer_start'] = answer_start
df.head()
```

```
data = df[['titles','context','answer_start','text','ids','is_impossible','questions']]
data.head()
```

```
import re
```

```
def noun_words(word):
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}
    return tag_dict.get(tag, wordnet.NOUN)
```

```
def pipeline(text):
  text = text.rstrip()
  stop_free = ' '.join([word for word in text.lower().split() if ((word not in stopword))])
  punc_free=re.sub('[^a-zA-Z]', " ", str(stop_free))
  text = ' '.join(LEM.lemmatize(word, noun_words(word)) for word in nltk.word_tokenize(punc_free))
  return text
```

```
from nltk.corpus import stopwords,wordnet
from nltk.tokenize import word_tokenize
```

```
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
```

```
stopword = set(stopwords.words('english'))
```

```
from nltk.stem import WordNetLemmatizer
LEM = WordNetLemmatizer()
```

```
data['clean_question'] = data['questions'].apply(pipeline)
df.head()
```
